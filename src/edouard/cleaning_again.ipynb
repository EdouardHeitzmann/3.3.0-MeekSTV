```python
from votekit.pref_profile import PreferenceProfile
from votekit.cvr_loaders import load_csv, load_ranking_csv
from votekit.cleaning import remove_repeat_cands_rank_profile, remove_and_condense_rank_profile

Portland_D4 = load_ranking_csv("/home/nardo/research/contribute/3.3.0-MeekSTV/data/portland/Portland_D4_raw_votekit_format.csv", rank_cols= [1,2,3,4,5,6])


mama_this_is_garbaj = [key for key in Portland_D4.candidates if str(key).startswith("Write-in")]
mama_this_is_garbaj += [key for key in Portland_D4.candidates if str(key).startswith("Uncertified")]
mama_this_is_garbaj += [key for key in Portland_D4.candidates if str(key).startswith("rank")]
mama_this_is_garbaj += [key for key in Portland_D4.candidates if str(key).startswith("Rank")]
mama_this_is_garbaj += ["overvote", "undervote", "skipped"]

new_profile = remove_repeat_cands_rank_profile(Portland_D4)
new_profile = remove_and_condense_rank_profile(mama_this_is_garbaj, new_profile)
new_profile = new_profile.group_ballots()

new_profile
```
